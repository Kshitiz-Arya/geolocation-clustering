```python
import folium
def plot(polygons:list, map):
     for p in polygons:
          folium.vector_layers.Polygon(locations=p , stroke=True, color='#ff00ff', weight=10
     ).add_to(map)
     return map
```

```python
m = folium.Map(location= (32.14403172857311, 34.80008052431097), position='relative')

# extract polygon from cluster and call plot with it
for c in cluster.values():
    polygon, time = c.values()
    folium.vector_layers.Polygon(locations=polygon[0], color='red', opacity=0.7, weight=7, tooltip=str(time)).add_to(m)

m
```
